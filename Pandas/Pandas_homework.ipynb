### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv(f'H:/studing/ML/продв. инж. данных/10._Основы_pandas/power.csv')

filtered_countryes = power[ (power['country'] == 'Latvia') | (power['country'] == 'Lithuania') | (power['country'] == 'Estonia') ]
filtered_categ = filtered_countryes[ (filtered_countryes['category'] == 4) | (filtered_countryes['category'] == 12) | (filtered_countryes['category'] == 21) ]
filtered_year = filtered_categ[ (filtered_categ['year'] >= 2005) | (filtered_categ['year'] <= 2010) ]
filtered_q = filtered_year[ (filtered_year['quantity'] > 0) ]

counter = 0
for i in filtered_q['quantity']:
    counter += i
print(counter)
```
